```python
from sklearn.impute import SimpleImputer
import pandas as pd 
import numpy as np
```

```python
#Mean
data = {'Marks': [55,75,80,np.nan,85,15,np.nan,13,20]}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='mean')
df[['Marks']] = impute.fit_transform(df[['Marks']])
print('After Impute', df)
```

```python
# Median
data = {'Marks': [80,45,np.nan,50,74,25,93,np.nan,33,23,7,17,65]}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='median')
df[['Marks']] = impute.fit_transform(df[['Marks']])
print('After Impute', df)
```

```python
# Most frequent (number)
data = {'Marks': [60,86,56,56,56,np.nan,65,34,34,23,21,np.nan,43,np.nan]}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='most_frequent')
df[['Marks']] = impute.fit_transform(df[['Marks']])
print('After Impute', df)
```

```python
# Most frequent (string)
data = {'cities': ['mecca','paris','ayodhya','masachusits',np.nan, 'masachusits']}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='most_frequent')
df[['cities']] = impute.fit_transform(df[['cities']])
print('After Impute', df)
```

```python
# constant
data = {'Marks': [60,86,56,56,56,np.nan,65,34,34,23,21,np.nan,43,np.nan]}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='constant',fill_value=1)
df[['Marks']] = impute.fit_transform(df[['Marks']])
print('After Impute', df)
```

```python
# constant (string)
data = {'cities': ['mecca','paris','ayodhya','masachusits',np.nan, 'masachusits']}
df = pd.DataFrame(data)
print('Data before impute', df)
impute = SimpleImputer(strategy='constant',fill_value='mozambique')
df[['cities']] = impute.fit_transform(df[['cities']])
print('After Impute', df)
```
