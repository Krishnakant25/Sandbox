```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, recall_score, f1_score, r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\python assignments\Student_Performance.csv')
df.head()
```

```python
x = df[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
y = df['Performance Index']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3, random_state= 42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
print(f"model intercept: {model.intercept_}")
print(f"model coeffecient: {model.coef_}")
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(x_test, y_test)
```
