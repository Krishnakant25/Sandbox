```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\python assignments\Ice_cream selling data.csv')
df.head()
```

```python
x = df[['Temperature (°C)']]
y = df['Ice Cream Sales (units)']
```

```python
df[['Temperature (°C)', 'Ice Cream Sales (units)']].plot(ax=plt.gca())
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
```

```python
poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(x_train)
```

```python
model.fit(x_poly,y_train)
```

```python
x_new = poly.transform(x_test)
y_new_pre = model.predict(x_new)
```

```python
mean_absolute_error(y_test,y_new_pre)
```

```python
mean_squared_error(y_test,y_new_pre)
```

```python
r2_score(y_test,y_new_pre)
```
